```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_pickle('heart_df.pkl')
```

```python
df.head()
```

```python
from sklearn.preprocessing import OrdinalEncoder
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
```

```python
from sklearn.model_selection import train_test_split
from sklearn.model_selection import learning_curve
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline
```

```python
from sklearn.metrics import accuracy_score,recall_score,confusion_matrix,classification_report
```

```python
cat_cols = df.select_dtypes('object').columns.tolist()
```

```python
X= df.iloc[:,:-1]
y = df.iloc[:,-1]
```

```python
X_train,X_test,y_train,y_test = train_test_split(X,y,random_state=42,test_size = 0.2)
```

```python
preprocessor = ColumnTransformer([
    ('trf',OrdinalEncoder(),cat_cols)
])
```

```python
pipe = Pipeline( steps=[
    ('step1',preprocessor),
    ('step2',XGBClassifier(max_depth=3,n_estimators=200,learning_rate=0.3))
])
```

```python
pipe.fit(X_train,y_train)
```

```python
y_pred = pipe.predict(X_test)
```

```python
print('Accuracy :  ',accuracy_score(y_test,y_pred))
print(classification_report(y_test,y_pred))
cm = confusion_matrix(y_test,y_pred)
sns.heatmap(cm , annot=True, fmt='d')
```

```python
# Checking on the train data 
y_pred = pipe.predict(X_train)
print('Accuracy :  ',accuracy_score(y_train,y_pred))

print(classification_report(y_train,y_pred))
cm = confusion_matrix(y_train,y_pred)
sns.heatmap(cm , annot=True, fmt='d')
```

I wanted the recall to be between 80-90 i am getting those results.

Recall is much important in datasets like this

```python
import pickle
pickle.dump(pipe,open('pipe.pkl','wb'))
```
